## 3. For each categorical variable, print frequency and relative frequency tables.

次數表:

```python
xCategorical = [
 'workclass',
 'education',
 'marital-status',
 'occupation',
 'relationship',
 'race',
 'sex',
 'native-country',
 '50K']
for cate in xCategorical:    
    print(cate)
    xtemp = adult_data[cate]
    print(xtemp.value_counts()) 
    print('\n')
```

相對次數表:

```python
for cate in xCategorical:    
    print(cate)
    xtemp = adult_data[cate]
    print(xtemp.value_counts(normalize=True)) 
    print('\n')
```

## 4. Draw bar plots for each categorical variable .
並且得到每個變數的長條圖

```python
import matplotlib.pyplot as plt
times = 0
plt.figure(figsize=(15, 15))
for cate in xCategorical: 
    times += 1
    plt.subplot(3,3,times)
    plt.tight_layout()
    adult_data[cate].value_counts().plot.bar(title='Freq dist of ' + cate)
```

## 5. For each continuous variable, draw histograms and distribution plots.
隨後取得連續型變數的直方圖與分布圖。

```python
xContinue = ['age',
 'fnlwgt',
 'education-num',           
 'capital-gain',
 'capital-loss',
 'hours-per-week']
for cont in xContinue:  
    adult_data[cont].hist(bins= 20) 
    plt.title(cont)
    plt.show()
```

```python
import seaborn as sns
fig, ax = plt.subplots(len(xContinue), figsize=(16,12))
for i, col_val in enumerate(xContinue):
    sns.distplot(adult_data[col_val], hist = True, ax = ax[i])
    ax[i].set_title('Freq dist '+col_val, fontsize=10)
    ax[i].set_xlabel(col_val, fontsize = 8)
    ax[i].set_ylabel('Count', fontsize = 8)
plt.tight_layout()
plt.show()
```

## 6. Draw pairwise scatterplots for continuous variables. Explain any discernable patterns?
可見兩種資料中是以比較平均的分布出現，pairplot的點基本都是照著原先變數分布的特性呈現，較難捕捉變數之間相關的地方，但還是有些可觀察的點:
